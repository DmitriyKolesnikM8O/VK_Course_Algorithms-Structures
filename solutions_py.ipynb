Задача 1.3.1: массив с числами, нужно 0 в конец

```python
n = int(input())
numbers = list(map(int, input().split()))

result = [x for x in numbers if x != 0] + [0] * numbers.count(0)

print(" ".join(map(str, result)))
```

Задача 1.5: два соседних числа из списка с наименьшей абсолютной разницей

```python
n = int(input())
numbers = list(map(int, input().split()))

one_ids = 0
two_ids = 1
for i in range(len(numbers) - 1):
  if abs(numbers[i] - numbers[i + 1]) < abs(numbers[one_ids] - numbers[two_ids]):
    one_ids = i
    two_ids = i + 1

print(f"{numbers[one_ids]} {numbers[two_ids]}")
```

Задача 1.7: последнее четное число в массиве или -1, если нет

```python
n = int(input())
numbers = list(map(int, input().split()))

numbers = numbers[::-1]
elem = -1
for i in numbers:
  if i % 2 == 0:
    elem = i
    break
print(elem)
```

Задача 2.1: В массиве найти индекс числа или индекс, это число должно было бы быть

```python
n = int(input())
numbers = list(map(int, input().split()))
target = int(input())

numbers = sorted(numbers)
ids = len(numbers)
for i in range(len(numbers)):
  if numbers[i] >= target:
    ids = i
    break

print(ids)
```

Задача 3.2.1: число, которое встречается в массиве больше половины раз или -1 в противном

```python
n = int(input())
numbers = list(map(int, input().split()))
numbers = sorted(numbers)

dictionary = dict()

for i in numbers:
  if i in dictionary:
    dictionary[i] += 1
  else:
    dictionary[i] = 1

number = max(dictionary, key=dictionary.get)
if dictionary[number] > (len(numbers) / 2):
  print(number)
else:
  print(-1)
```

Задача 4.3: макс кол-во повторения 1 буквы в строке

```python
row = input()

dictionary = dict()

for i in row:
  if i in dictionary:
    dictionary[i] += 1
  else:
    dictionary[i] = 1

print(dictionary[max(dictionary, key=dictionary.get)])
```

Задача 5.1: максимальное произведение из подпоследовательности

```python
from itertools import permutations

n = int(input())
numbers = list(map(int, input().split()))
count = int(input())

permuts = list(permutations(numbers, count))
max_res = float("-inf")
for permut in permuts:
  res = 1
  for i in permut:
    res *= i
  if res > max_res:
    max_res = res

print(max_res)
```
