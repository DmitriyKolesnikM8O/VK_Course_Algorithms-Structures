Задача 1.3.1: массив с числами, нужно 0 в конец

```python
n = int(input())
numbers = list(map(int, input().split()))

result = [x for x in numbers if x != 0] + [0] * numbers.count(0)

print(" ".join(map(str, result)))
```

Задача 1.3.2: массив с числами + цифра для удаления, удалить все ее вхождения в массив

```python
n = int(input())
numbers = list(map(int, input().split()))
elem = int(input())

result = [x for x in numbers if x != elem]

print(" ".join(map(str, result)))
```

Задача 1.5: два соседних числа из списка с наименьшей абсолютной разницей

```python
n = int(input())
numbers = list(map(int, input().split()))

one_ids = 0
two_ids = 1
for i in range(len(numbers) - 1):
  if abs(numbers[i] - numbers[i + 1]) < abs(numbers[one_ids] - numbers[two_ids]):
    one_ids = i
    two_ids = i + 1

print(f"{numbers[one_ids]} {numbers[two_ids]}")
```

Задача 1.7: последнее четное число в массиве или -1, если нет

```python
n = int(input())
numbers = list(map(int, input().split()))

numbers = numbers[::-1]
elem = -1
for i in numbers:
  if i % 2 == 0:
    elem = i
    break
print(elem)
```
